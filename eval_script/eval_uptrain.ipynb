```python
import pandas as pd
from uptrain import EvalLLM, Evals
import json

# Create context data dictionary
json_file_path = "context_data.json"

# Initialize an empty dictionary
context_dict = {}

# Import the JSON file and create a dictionary
with open(json_file_path, 'r') as json_file:
    context_dict = json.load(json_file)

# Load the Excel file
excel_file = "eval_dataset.xlsx"
df = pd.read_excel(excel_file)
```

```python
import openai
import os
from dotenv import load_dotenv


load_dotenv()
openai.api_key = os.environ.get("OPENAI_API_KEY")
```

```python
# Define a function to evaluate each question-context-response from the excel file
def process_batch(start_row, end_row, result_variable):
    batch_data = []
    for _, row in df.iloc[start_row:end_row].iterrows():
        question = row['question']
        context = row['context']
        response = row['response']
        batch_data.append({"question": [question], "context": context_dict[context], "response": [response]})
    
    print(f"Evaluating rows {start_row} to {end_row}")
    
    # Initialize EvalLLM
    
    eval_llm = EvalLLM(openai_api_key=openai.api_key)
    
    # Evaluate the data
    result = eval_llm.evaluate(
        data=batch_data,
        checks=[Evals.CONTEXT_RELEVANCE, Evals.FACTUAL_ACCURACY, Evals.RESPONSE_COMPLETENESS]
    )
    
        # Save the evaluation result to a JSON file
    with open(result_variable, 'w') as f:
        json.dump(result, f, indent=4)
    
    print(f"Evaluation result saved to {result_variable}")
```

```python
process_batch(0, 5, 'abuse.json')
process_batch(5, 10, 'byod.json')
process_batch(10, 15, 'client_money.json')
process_batch(15, 20, 'client_right.json')
process_batch(20, 25, 'continence_mana.json')
process_batch(25, 30, 'continence_improve.json')
process_batch(30, 35, 'data_breach.json')
process_batch(35, 40, 'decision.json')
process_batch(40, 45, 'deterioration.json')
process_batch(45, 50, 'diversity.json')
process_batch(50, 55, 'fall.json')
process_batch(55, 60, 'feedback.json')
process_batch(60, 65, 'grievance.json')
process_batch(65, 70, 'handbook.json')
process_batch(70, 75, 'incident.json')
process_batch(75, 80, 'salary.json')
process_batch(80, 85, 'risk.json')
process_batch(85, 90, 'infection.json')
process_batch(90, 95, 'privacy.json')
```

```python
def combine_json(files, combined_file):
    combined_data = []
    
    # Read contents of each JSON file and append to combined_data
    for file in files:
        with open(file, 'r') as f:
            data = json.load(f)
            combined_data.extend(data)
    
    # Write the combined data to the new JSON file
    with open(combined_file, 'w') as cf:
        json.dump(combined_data, cf, indent=4)
    
    print(f"Combined JSON saved to {combined_file}")
```

```python
json_files = ['abuse.json','byod.json','client_money.json','client_right.json',
              'continence_mana.json','continence_improve.json','data_breach.json',
              'decision.json','deterioration.json','diversity.json','fall.json',
              'feedback.json','grievance.json','handbook.json','incident.json',
              'salary.json','risk.json','infection.json','privacy.json']
combine_json(json_files, 'all_result.json')
```

```python
import matplotlib.pyplot as plt
import json

# Load the evaluation results from the JSON file
with open('all_result.json', 'r') as f:
    evaluation_results = json.load(f)

# Define lists to store scores
context_relevance_scores = []
factual_accuracy_scores = []
response_completeness_scores = []

# Iterate over each evaluation result
for result in evaluation_results:
    # Extract scores
    context_relevance_scores.append(result['score_context_relevance'])
    factual_accuracy_scores.append(result['score_factual_accuracy'])
    response_completeness_scores.append(result['score_response_completeness'])

# Create bar plots for each type of score
plt.figure(figsize=(12, 5))

# Context Relevance
plt.subplot(1, 3, 1)
plt.hist(context_relevance_scores, bins=10, color='skyblue', edgecolor='black')
plt.title('Context Relevance Scores', fontsize=14)
plt.xlabel('Score', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

# Factual Accuracy
plt.subplot(1, 3, 2)
plt.hist(factual_accuracy_scores, bins=10, color='lightgreen', edgecolor='black')
plt.title('Factual Accuracy Scores', fontsize=14)
plt.xlabel('Score', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

# Response Completeness
plt.subplot(1, 3, 3)
plt.hist(response_completeness_scores, bins=10, color='salmon', edgecolor='black')
plt.title('Response Completeness Scores', fontsize=14)
plt.xlabel('Score', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()

# Export the plot as PNG
plt.savefig('evaluation_scores.png', dpi=300)  # Adjust dpi for better quality if needed
plt.show()
```

```python
import json
import csv
import numpy as np

# List to store results
results = []

# List of JSON files
json_files = ['abuse.json', 'byod.json', 'client_money.json', 'client_right.json',
              'continence_mana.json', 'continence_improve.json', 'data_breach.json',
              'decision.json', 'deterioration.json', 'diversity.json', 'fall.json',
              'feedback.json', 'grievance.json', 'handbook.json', 'incident.json',
              'salary.json', 'risk.json', 'infection.json', 'privacy.json']

# Iterate over each JSON file
for file in json_files:
    # Load JSON file
    with open(file, 'r') as f:
        data = json.load(f)
    
    # Extract scores
    context_relevance_scores = [entry['score_context_relevance'] for entry in data]
    factual_accuracy_scores = [entry['score_factual_accuracy'] for entry in data]
    response_completeness_scores = [entry['score_response_completeness'] for entry in data]
    
    # Calculate mean and standard deviation
    mean_context_relevance = np.mean(context_relevance_scores)
    std_dev_context_relevance = np.std(context_relevance_scores)
    
    mean_factual_accuracy = np.mean(factual_accuracy_scores)
    std_dev_factual_accuracy = np.std(factual_accuracy_scores)
    
    mean_response_completeness = np.mean(response_completeness_scores)
    std_dev_response_completeness = np.std(response_completeness_scores)
    
    # Append results
    results.append({
        'File': file,
        'Mean Context Relevance': mean_context_relevance,
        'Standard Deviation Context Relevance': std_dev_context_relevance,
        'Mean Factual Accuracy': mean_factual_accuracy,
        'Standard Deviation Factual Accuracy': std_dev_factual_accuracy,
        'Mean Response Completeness': mean_response_completeness,
        'Standard Deviation Response Completeness': std_dev_response_completeness,
        'Context Relevance Scores': context_relevance_scores,
        'Factual Accuracy Scores': factual_accuracy_scores,
        'Response Completeness Scores': response_completeness_scores
    })

# Write results to CSV file
csv_columns = ['File', 'Mean Context Relevance', 'Standard Deviation Context Relevance',
               'Mean Factual Accuracy', 'Standard Deviation Factual Accuracy',
               'Mean Response Completeness', 'Standard Deviation Response Completeness',
               'Context Relevance Scores', 'Factual Accuracy Scores', 'Response Completeness Scores']

csv_file = 'evaluation_results.csv'

with open(csv_file, 'w', newline='') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=csv_columns)
    writer.writeheader()
    for data in results:
        writer.writerow(data)

print("CSV file created successfully.")
```
